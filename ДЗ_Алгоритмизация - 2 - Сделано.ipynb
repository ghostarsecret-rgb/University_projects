```python
list1 = [6, 12, 4, 3, 8]

def sort_for(list1, info=False, method=str):
    list2 = []
    step = 0
    if info==True: print(f'Start list: {list1}')
    for i in range(len(list1)):
        if method=='min': peak = min(list1)
        if method=='max': peak = max(list1)
        list2.append(peak)
        list1.remove(peak)
        step += 1
        if info==True: print(list2)
    if info == True: print(f'Final result: {list2}\nSteps: {step}' )
    else: print(list2)
sort_for(list1, True, 'max')
```

```python
import random

def sort_while(list, info=False, method=str):
  sorted = []
  list = list.copy()
  step = 0
  if info==True: print(f'Start list: {list}')
  while len(list) != 0:
    if method=='min': peak = min(list)
    if method=='max': peak = max(list)
    sorted.append(peak)
    list.remove(peak)
    step += 1
    if info==True: print(sorted)
  if info==True: print(f'Final result: {sorted}\nSteps: {step}')
  else: print(sorted)

rand_list = [random.randint(0, 50) for x in range(10)]
sorted = sort_while(rand_list, True, 'max')
```

Решить задачу следующими способами:

```python
import random
def bubble_for(list, info=False, method=str):
  if info==True: print(f'Start list: {list}')
  if method=='min': method = 1
  if method=='max': method = -1
  step = 0
  for iter in range(len(list)):
    g = 0
    for i in range(len(list) - 1):
      if list[i] > list[i+1]:
        list[i], list[i+1] = list[i+1], list[i]
        g += 1
    if g == 0:
      break
    if info==True: print(list)
    step += 1
  if info==True: print(f'Final result: {list[::method]}\nSteps: {step}')
  else: print(list)

main_massive = [random.randint(0, 10) for x in range(15)]
bubble_for(main_massive, info=True, method='max')
```

```python
import random
def bubble_while(list, info=False, method=str):
  i = 0
  op_count = 0
  step = 0
  if method=='min': method = 1
  if method=='max': method = -1
  print(f'Start list: {list}')
  while True:
    if list[i] > list[i+1]:
      list[i], list[i+1] = list[i+1], list[i]
      op_count += 1
    i += 1
    if i == len(list)-1:
      if op_count == 0:
        break
      i = 0
      op_count = 0
    if info==True: print(list)
    step += 1
  if info==True: print(f'Final result: {list[::method]}\nSteps: {step}')
  else: print(list)

main_massive = [random.randint(0, 10) for x in range(10)]
bubble_while(main_massive, info=True, method='max')
```

  **Задача № 9**

Имеется отсортированный массив целых чисел L1, состоящий из 50 элементов в диапазоне значений от 0 до 100.

Необходимо осуществить поиск элемента со значением 10.
Определить итоговую сложность предложенной программы.


```python
import math
import random
def bin_search(list, need):
  step = math.floor(math.log(len(list)//2))
  middle = len(list)//2
  print(f'Start list: {list}')
  while True:
    if need not in list:
      print('Theres is no such number')
      break
    if list[middle] < need:
      middle += 2**step
      print(list[middle])
    else: middle -= 2**step
    print(list[middle])
    if step <= 0:
      step = 1
    if list[middle] == need and list[middle-1] < need:
      break
    step -= 1
  if need in list: print(f"Поиск завершён\n Искомое - {need}\n Позиция - {middle+1}")


rand = sorted([random.randint(0,100) for x in range(50)])
bin_search(rand, 10)
```

Итоговая сложность
- Теоретическая сложность: O(log n)
- Практическая сложность данного кода из-за особенностей реализации: до O(n)
- Следует оптимизировать реализацию для классического бинарного поиска и избегать дорогостоящих проверок внутри цикла.

Определить итоговую сложность предложенной программы.




```python
import math
import random
def bin_search(list, need):
  step = math.floor(math.log(len(list)//2))
  middle = len(list)//2
  print(f'Start list: {list}')
  while True:
    if need not in list:
      print('Theres is no such number')
      break
    if list[middle] < need:
      middle += 2**step
      print(list[middle])
    else: middle -= 2**step
    print(list[middle])
    if step <= 0:
      step = 1
    if list[middle] == need and list[middle-1] < need:
      break
    step -= 1
  if need in list: print(f"Поиск завершён\n Искомое - {need}\n Позиция - {middle+1}")


def sorting_one(list, info=False, method=str):
  print(f'Start list: {list}')
  step = 0
  for ind in range(len(list)):
    if method=='min': peak = min(list[ind:])
    if method=='max': peak = max(list[ind:])
    mini = list[ind:].index(peak)+ind
    if list[ind] != list[mini]: list[mini], list[ind] = list[ind], list[mini]
    else: continue
    if info==True: print(list)
    step += 1

  if info==True: print(f'Final result: {list}\nSteps: {step}')
  else: print(list)


rand = [random.randint(0,100) for x in range(50)]
rand = sorting_one(rand, False, 'min')
bin_search(rand, 10)
```

Вывод:

sorting_one: O(n²) — Selection Sort с поиском минимума/максимума на каждом шаге  
bin_search (текущая реализация): до O(n) из-за полного сканирования списка на каждом шаге  
Рекомендация для bin_search: классический бинарный поиск — O(log n)
